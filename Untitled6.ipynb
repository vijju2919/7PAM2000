```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
dframe = pd.read_csv("world-happiness-report-2021.csv")
```

```python
dframe.hist(column='Ladder score')
```

```python
data = pd.read_csv('world-happiness-report-2021.csv')
x = data ['Logged GDP per capita']
y = data['Ladder score']
```

```python
plt.scatter(x, y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
x = data['Logged GDP per capita'].values.reshape(-1,1)
y = data['Ladder score']
lr_model = LinearRegression()
lr_model.fit(x, y)
y_pred = lr_model.predict(x)
plt.scatter(x, y)
plt.xlabel('Logged GDP per capita')
plt.ylabel('Ladder score')
plt.plot(x, y_pred)
```

```python
theta_0 = lr_model.intercept_
theta_1 = lr_model.coef_
theta_0, theta_1
```
